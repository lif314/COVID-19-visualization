## urllib

- request.urlopen()

```python
from urllib import request
```

```python
url = "http://www.baidu.com"
res = request.urlopen(url)
```

```python
print(res.info())  # 
```

```python
print(res.getcode())  # 
```

```python
print(res.geturl())  # url
```

```python
html = res.read()  #   
html = html.decode('utf-8')   # 
print(html)
```

- User-Agent

```python
url = "https://www.dianping.com/"

# User-Agent 
header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.63 Safari/537.36"
}

req  = request.Request(url, headers=header)
res = request.urlopen(req)
print(res.info())
```

## beautifulsoup4

- BeautifulSoup(html)



```python
import requests

from bs4 import BeautifulSoup
```

```python
url = "http://wsjkw.sc.gov.cn/scwsjkw/gzbd/fyzt.shtml"

res = requests.get(url)

res.headers
```

```python
res.encoding = 'utf-8'
html = res.text
# print(html)
```

```python
soup = BeautifulSoup(html)
soup.find('h2').text
```

```python
# a
a =  soup.find('a')
print(a)
```

```python
# a
print(a.attrs)
```

```python
# herf
print(a.attrs['href'])
```

```python
url_new = "http://wsjkw.sc.gov.cn" + a.attrs['href']
print(url_new)
```

```python
resp = requests.get(url_new)

print(resp.status_code)
```

```python
resp.encoding = 'utf=8'
html = resp.text
# print(html)
soup = BeautifulSoup(html)
context = soup.find_all('p')[1]

print(type(context))
print(context)
```

## re



```python
import re

text = context.text

pattern = "(\d+)"

res = re.search(pattern, text)

print(res)
```

```python
print(res.group()) # 
print(res.group(0)) #  
print(res.group(1)) # 2
```

```python
# 532
# 156

# .*    .*? 
pattern1 = "(\d+).*?(\d+).*?(\d+)"

res = re.search(pattern1, text)

print(res)
```

```python
print(res.groups())
```
